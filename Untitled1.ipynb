```python
scaler=MinMaxScaler()
train=train_df.drop(columns=['LoanID','Default'],axis=1)
train_normalized=scaler.fit_transform(train)
train_normalized_df=pd.DataFrame(train_normalized,columns=train.columns)
train_normalized_df.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score, roc_auc_score,roc_curve
from sklearn.model_selection import train_test_split as tts
```

```python
X=train_normalized_df
y=train_df['Default']
X_train,X_test,y_train,y_test=tts(X,y,test_size=0.2,random_state=2)
```

```python
X_train.head()
```

```python
classifiers=[LogisticRegression,DecisionTreeClassifier,RandomForestClassifier,GradientBoostingClassifier]
plt.figure(figsize=(8,20))
for i,classifier in enumerate(classifiers,1):
    classify=classifier(random_state=42)
    classify.fit(X_train,y_train)
    y_pred=classify.predict(X_test)
    y_pred_prob=classify.predict_proba(X_test)[:,1]
    accuracy=accuracy_score(y_test,y_pred)
    fpr,tpr,_=roc_curve(y_test,y_pred_prob)
    roc_auc = roc_auc_score(y_test, y_pred_prob)
    print(f"Accuracy Score for {classifier} is {accuracy} ")
    print()
    print(f'Roc Accuracy for {classifier} is {roc_auc}')
    print()
    plt.subplot(len(classifiers),1,i)
    plt.plot(fpr,tpr,color='darkorange',lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'Receiver Operating Characteristic for {classifier}')
    plt.legend(loc='lower right')
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import GradientBoostingClassifier
from scipy.stats import uniform, randint

# Define the parameter distribution
param_dist = {
    'n_estimators': randint(100, 300),
    'learning_rate': uniform(0.01, 0.2),
    'max_depth': randint(3, 7),
    'min_samples_split': randint(2, 11),
    'min_samples_leaf': randint(1, 5),
    'subsample': uniform(0.8, 0.2)
}

# Initialize the model
gbc = GradientBoostingClassifier()

# Initialize Randomized Search
random_search = RandomizedSearchCV(estimator=gbc, param_distributions=param_dist, n_iter=50, cv=3, n_jobs=-1, verbose=2, scoring='accuracy', random_state=42)

# Fit Randomized Search
```

```python
random_search.fit(X_train, y_train)
```

```python
 #Best model
best_gbc = random_search.best_estimator_
# Make predictions
y_pred = best_gbc.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
y_pred_prob = best_gbc.predict_proba(X_test)[:, 1]
roc_auc = roc_auc_score(y_test, y_pred_prob)

print(f'Improved GradientBoostingClassifier Accuracy: {accuracy}')
print(f'Improved GradientBoostingClassifier ROC-AUC Score: {roc_auc}')
```
